### Average Sugar by Beverage Category

```python
d1 = df.groupby(['Beverage_category']).mean()
```

```python
d1.sort_values(' Sugars (g)', ascending = False).plot.barh(y=' Sugars (g)', rot=0)
plt.title('Average Sugars by Beverage Category')
plt.xlabel("Beverage Category", labelpad=14)
plt.ylabel("Sugars (g)", labelpad=14)
```

From the above graph, we can see that Coffee has zero sugar, Frappuccino drinks have an average of over 50 grams of sugar per drink. If you wonder what it means, try eating 50 grams of sugar (that's around 12 teaspoons!)

Let's take a look at sugar amount for each beverage

```python
d2 = df.groupby(['Beverage']).mean()
d2.sort_values(' Sugars (g)', ascending = False).head()
```

```python
d2.sort_values(' Sugars (g)', ascending = False).plot.bar(y=' Sugars (g)', rot=0, cmap="plasma", alpha=0.4,)
plt.title('Average Sugars by Beverage', fontsize = 15)
plt.xlabel("Beverage", labelpad=10)
plt.ylabel("Sugars (g)", labelpad=10)
plt.xticks(rotation = 90)
```

We can see that all drinks that fall under frapuccino category are high in sugar. Java Chip Without Whipped Cream has, on average, 62 grams of sugar per drink! Let's take a look at the Daily Value. According to the American Heart Association (AHA), the recommended added sugar intake is *no more than 37.5g* per day for men and *no more than 25g* per day for women.

So, we divide the sugars in each of the drink by 37.5 (for men) or 25 (for women), and multiply it with 100, to obtain the percentage.

```python
d3 = df.groupby(['Beverage']).mean()/37.5*100
d3.sort_values(' Sugars (g)', ascending = False).head()
```

```python
d3.sort_values(' Sugars (g)', ascending = False).plot.bar(y=' Sugars (g)', rot=0, color="Green")
plt.title('Daily Suggested Value: Sugar in Percentage', fontsize = 15)
plt.xlabel("Beverage", labelpad=10)
plt.ylabel("Daily Suggested Value %", labelpad=10)
plt.xticks(rotation = 90)
```

Wow, that's 160% of daily suggested value for a single Java Chip Frap on average!

What about for women?


```python
d4 = df.groupby(['Beverage']).mean()/25*100
d4.sort_values(' Sugars (g)', ascending = False).head()
```

```python
d4.sort_values(' Sugars (g)', ascending = False).plot.bar(y=' Sugars (g)', rot=0, color="Green")
plt.title('Daily Suggested Value: Sugar in Percentage', fontsize = 15)
plt.xlabel("Beverage", labelpad=10)
plt.ylabel("Daily Suggested Value %", labelpad=10)
plt.xticks(rotation = 90)
```

Almost 250% of daily suggested value in a single Java Chip Frap on average for a woman, isn't that scary?

People may wonder what happens if you take too much sugar?

- Gain weights
- Increase risk of heart diseases
    - A study in over 30,000 people found that those who consumed 10-25% of calories from added sugar had a 38% greater risk of dying from heart disease, compared to those consuming less than 10% of calories from added sugar
- Increase risk of developing acne
    - A study conducted on 2300 teens found that those who frequently consumed added sugar had a 30% greater risk of developing acne
- Increase risk of type 2 diabetes
- Increase cellular aging, skin aging process
- And so on...


So, no Starbucks any more? What if I'm a Starbucks lover who cannot live without Starbucks?

Well, there are some "safe" drinks that have, on average, less than 30 grams of added sugar.

```python
d5 = df.groupby(['Beverage']).mean()
d5 = d5[d5[' Sugars (g)'] < 30]
d5.sort_values(' Sugars (g)', ascending = False).plot.barh(y=' Sugars (g)', rot=0, cmap="plasma", alpha=0.4,)
plt.title('Average Sugars by Beverage', fontsize = 15)
plt.xlabel("Beverage", labelpad=10)
plt.ylabel("Sugars (g)", labelpad=10)
#plt.xticks(rotation = 90)
```

How about cup sizes? The smaller the cup size, lesser added sugar you will consume, is that right?

```python
d6 = df.groupby(['Beverage_prep']).mean()
#d6 = d5[d5[' Sugars (g)'] < 30]
d6.sort_values(' Sugars (g)', ascending = False).plot.barh(y=' Sugars (g)', rot=0)
plt.title('Average Sugars by Beverage Size', fontsize = 15)
plt.xlabel("Beverage", labelpad=10)
plt.ylabel("Sugars (g)", labelpad=10)
#plt.xticks(rotation = 90)
```

From the chart above, we can tell that the smaller the cup size, the lesser sugar you will eat on average. Also note that, adding extra milk can mean adding extra sugar because of lactose. Whole milk has on average over 55 grams of sugar while soymilk has only 30 grams (still a lot, but significantly lesser). If you must add milk to your coffee, try nonfat milk or soymilk.
